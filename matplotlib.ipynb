```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import pandas as pd
```

## Generating cos and sin graph using matplotlib

```python
x = np.linspace(0 , 4*np.pi , 100)
y = np.linspace(0 , 4*np.pi , 100)
```

```python
sin_y = np.sin(x)
cos_y = np.cos(y)
```

```python
plt.plot(x , sin_y, label = "Sine")
plt.plot(y , cos_y , label = "Cosine")
plt.title("Sine Wave")
plt.legend()
plt.show()
```

```python
plt.style.available
```

## using different styles and saving the plot

```python
plt.style.use( 'bmh')
fig = plt.figure()
plt.plot(x , sin_y, label = "Sine")
plt.plot(y , cos_y , label = "Cosine")
plt.title("Sine  and Cos Wave")
plt.legend()
plt.figure(figsize=(8,15))
fig.savefig("sine and cos.png")
plt.show()
```

## DIFFERENT TYPES OF PLOTS
###LINE PLOT

single line plot

```python
plt.style.use('Solarize_Light2')
x = np.linspace(0 , 4*np.pi , 50)
y = np.sin(x)
plt.plot(x,y)
plt.title("Sine  And Cos Wave")
# plt.legend()
plt.show()
```

Multiple line plot

```python
x = np.linspace(0 , 4*np.pi , 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.style.use('Solarize_Light2')
plt.figure(figsize=(15,5))

plt.plot(x , sin_y, label = "Sine" , color = 'red' , linewidth = 1)
plt.plot(x, cos_y , label = "Cosine" , color = 'blue' , linewidth = 1)

plt.title("Sine  and Cos Wave")

```
